# Finding Optimum number of Clusters By Elbow Method

```python
from sklearn.cluster import KMeans
x = df.iloc[:, [0,1,2,3]].values
print(x)
```

- In order to choose the best optimal value of K for the clustering we use a method called the elbow method.In this method we plot the WCSS(with in cluster sum of square ) which is the sum of square of all the points within a cluster and its centroid.

- You will se that as the number of cluster increases the WCSS will decrease but thier wil be an elbow point where there will be major change which you will select as K value and after that your value of WCSS will be almost parallel to the X-axis

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,init='k-means++', n_init=10, max_iter=300)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)
    print('k:',i ,"Error:",kmeans.inertia_) # kmeans.inertia_ returns the WCSS value for an initialized cluste
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

# Applying K Mean Clustering Algorithm

- from the above gtaph we can see that the elbow shape is formed at K = 3 so we will chosse K = 3 as the optimal value for the clustering

```python
# applying K mean clustering with a value of K =3

kmeans3 = KMeans(n_clusters=3)
y_kmeans = kmeans3.fit_predict(x)  #kmeans.fit_predict method returns the array of cluster labels each data point belongs to.
print(y_kmeans)  # it returns the total clusters



kmeans3.cluster_centers_
```

```python
#Ploting the scatter plots of every class (2 feaures at the same time)

#filter rows of original data
filtered_y_kmeans0 =x[y_kmeans == 0]
#print(filtered_y_kmeans0)

filtered_y_kmeans1= x[y_kmeans == 1]
#print(filtered_y_kmeans1)

filtered_y_kmeans2 =x[y_kmeans == 2]
 
#Plotting the results
plt.figure(figsize=(10,10))
plt.scatter(filtered_y_kmeans0[:,0] ,filtered_y_kmeans0[:,1] ,s =100, color = 'red', label='Iris-setosa')
plt.scatter(filtered_y_kmeans1[:,0] ,filtered_y_kmeans1[:,1] ,s =100, color = 'blue',label='Iris-versicolour')
plt.scatter(filtered_y_kmeans2[:,0] ,filtered_y_kmeans2[:,1] ,s =100, color = 'green',label='Iris-virginica')
plt.title('Iris Flower Clusters')
plt.xlabel('Sepal Length in cm')
plt.ylabel('Sepal Width in cm')
plt.legend()
plt.show()
```
